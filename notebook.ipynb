## Visualization & Explanatory Analysis

### Pertanyaan 1: Musim Apa Dengan Pelanggan Penyewa Sepeda Terbanyak?

```python
# Membuat grafik batang penyewaan sepeda per musim dengan warna yang ditentukan
colors = ["#FF6347", "#4682B4", "#32CD32", "#FFD700"]

fig, ax = plt.subplots(figsize=(10, 5))

sns.barplot(
    y="count", 
    x="season",
    data=day_df.sort_values(by="season", ascending=False),
    palette=colors,
    ax=ax,
    edgecolor='black',
    linewidth=1
)

ax.set_title("Grafik Penyewaan Sepeda Tiap Musim", fontsize=20)
ax.set_ylabel(None)
ax.set_xlabel(None)
ax.tick_params(axis='x', labelsize=20)
ax.tick_params(axis='y', labelsize=10)

plt.show()
```

```python
# Agregasi data per musim
season_data = day_df.groupby(by='season').agg({
    'count': 'sum'
}).reset_index()

season_data.columns = ['season', 'total_count']

# Temukan nilai maksimum dan musim terkait
max_count = season_data['total_count'].max()
max_season = season_data.loc[season_data['total_count'].idxmax(), 'season']

# Mapping musim ke nama
season_mapping = {
    1: 'Spring',
    2: 'Summer',
    3: 'Fall',
    4: 'Winter'
}

# Ubah musim numerik menjadi nama
max_season_name = season_mapping.get(max_season, 'Unknown')

print(f"Total penyewaan tertinggi adalah {max_count} pada musim {max_season_name}.")
```

### Pertanyaan 2: Pada Jam Berapa Total Penyewaan Sepeda Tertinggi Terjadi?

```python
# Mengelompokkan data berdasarkan 'hour' dan menghitung total penyewaan ('count')
hourly_data = hour_df.groupby(by='hour').agg({
    'count': ['sum']
})

hourly_data.columns = ['count_sum']
hourly_data = hourly_data.reset_index()

# Membuat plot garis tiap jam
plt.figure(figsize=(10, 6))
plt.plot(hourly_data['hour'], hourly_data['count_sum'], color='skyblue', marker='o')

plt.title('Total Count per Hour', fontsize=16)
plt.xlabel('Hour', fontsize=14)
plt.ylabel('Count', fontsize=14)

plt.xticks(ticks=hourly_data['hour'], labels=hourly_data['hour'])

plt.grid(True, axis='x', linestyle='--', linewidth=1)

plt.show()
     
```

```python
# Grouping data berdasarkan 'hour' dan menjumlahkan 'count'
hourly_data = hour_df.groupby(by='hour').agg({
    'count': ['sum']
})

hourly_data.columns = ['count_sum']
hourly_data = hourly_data.reset_index()

# Mencari nilai maksimum dan jam di mana itu terjadi
max_count = hourly_data['count_sum'].max()
max_hour = hourly_data['hour'][hourly_data['count_sum'].idxmax()]

print(f"Total penyewaan sepeda tertinggi adalah {max_count} pada jam {max_hour}.")
```

Berdasarkan analisis data penyewaan sepeda per jam, total penyewaan sepeda tertinggi terjadi pada jam 17. Pada jam ini, jumlah penyewaan sepeda mencapai nilai maksimum sebesar 336,860 unit.
